## BTS-Airline-DataCleaning

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/rochanehurst/Documents/Personal Projects/BTS - Airline Analysis/T_ONTIME_REPORTING.csv')
df
```

```python
# Setting delay columns as 0 when value is NaN
delay_cols = ["CARRIER_DELAY", "WEATHER_DELAY", "NAS_DELAY", "SECURITY_DELAY", "LATE_AIRCRAFT_DELAY"]
df[delay_cols] = df[delay_cols].fillna(0)
df
```

```python
df.drop_duplicates()
```

```python
df.loc["ORIGIN_CITY_NAME"] = df["ORIGIN_CITY_NAME"].str.strip()
df
```

```python
df.loc["DEST_CITY_NAME"] = df["DEST_CITY_NAME"].str.strip()
df
```

```python
df = df.fillna('')
df
```

```python
# Make all column names lowercase and snake_case
df.columns = (
    df.columns.str.strip()       # remove leading/trailing spaces
              .str.lower()       # lowercase
              .str.replace(' ', '_')  
              .str.replace('-', '_')  
)
```

```python
#Rename Columns
#The raw BTS dataset has long, uppercase column names.  
#I converted them to lowercase snake_case so they’re easier to reference in Pandas.
rename_dict = {
    'op_unique_carrier': 'carrier',
    'fl_date': 'flight_date',
    'origin_city_name': 'origin_city',
    'dest_city_name': 'dest_city',
    'arr_delay': 'arrival_delay',
    'dep_delay': 'departure_delay',
    'late_aircraft_delay': 'late_aircraft',
}
df = df.rename(columns=rename_dict)

# Check new column names
print(df.columns.tolist())
```

```python
df
```

```python
#convert flight_date from object into datetime
df['flight_date'] = pd.to_datetime(df['flight_date'])
```

```python
#convert cancelled and diverted into boolean types for easier analysis
df['cancelled'] = df['cancelled'].astype(bool)
df['diverted'] = df['diverted'].astype(bool)
```

```python
print(df.isna().sum)
```

```python
#Day of the week names instead of numbers
import calendar

# convert safely to numeric
df['day_of_week'] = pd.to_numeric(df['day_of_week'], errors='coerce')

# drop NaN values if any
df = df.dropna(subset=['day_of_week'])

# convert to int
df['day_of_week'] = df['day_of_week'].astype(int)

# map to week day names
df['day_of_week'] = df['day_of_week'].apply(lambda x: calendar.day_name[x-1])
```

```python
df
```

```python
#convert arrival_delay to numeric
df['arrival_delay'] = pd.to_numeric(df['arrival_delay'], errors = 'coerce')

# categorized types of delays

def delay_category(x):
    if x < 0:
        return "Early"
    elif 0 <= x <= 15:
        return "On Time / Minor"
    elif 15 <= x <= 60:
        return "Delayed"
    else:
        return "Severe Delay"
df['delay_category'] = df['arrival_delay'].apply(delay_category)
```

```python
df.to_csv("bts_airline_clean.csv", index = False)
```
